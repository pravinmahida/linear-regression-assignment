## Data Preparation

From above scatter plots and also the heat map as we can notice there is a high corelation between: 
carlength, curbweight, wheelbase and carwidth, so we can drop 3 out of 4,
so lets drop carwidth and curbweight and wheelbase 2.

There is a high corelation of .97 between highwaympg and citympg, so lets drop highwaympg



```python
df_carPrices.drop(['carwidth','curbweight','wheelbase','highwaympg'], axis =1, inplace = True)
#we can also remove carID  as its just a serial number 
df_carPrices.drop(['car_ID'], axis =1, inplace = True)
```

```python
df_carPrices.info()
```

```python
# Seperate Brand and Model from Car name
brand_model = df_carPrices["CarName"].str.split(" ", n = 1, expand = True)
df_carPrices["company"] = brand_model[0]
```

```python
# To Fix redundant Data. 
df_carPrices.CarName = df_carPrices['company'].str.lower()
df_carPrices['company'] = df_carPrices['company'].str.replace('alfaromeo','alfa')
df_carPrices['company'] = df_carPrices['company'].str.replace('vw','volkswagen')
df_carPrices['company'] = df_carPrices['company'].str.replace('vokswagen','volkswagen')
df_carPrices['company'] = df_carPrices['company'].str.replace('toyouta','toyota')
df_carPrices['company'] = df_carPrices['company'].str.replace('porcshce','porsche')
df_carPrices['company'] = df_carPrices['company'].str.replace('maxda','mazda')
df_carPrices['company'] = df_carPrices['company'].str.replace('maxda','mazda')
df_carPrices['company'] = df_carPrices['company'].str.replace('alfa-romero','alfa')
df_carPrices['company'] = df_carPrices['company'].str.replace('Nissan','nissan')

# list the company names.
df_carPrices.company.unique()
```

```python
# Converting Yes to 1 and No to 0
df_carPrices['fueltype'] = df_carPrices['fueltype'].map({'gas': 1, 'diesel': 0})
df_carPrices['aspiration'] = df_carPrices['aspiration'].map({'std': 1, 'turbo': 0})
df_carPrices['doornumber'] = df_carPrices['doornumber'].map({'two': 1, 'four': 0})
df_carPrices['enginelocation'] = df_carPrices['enginelocation'].map({'front': 1, 'rear': 0})

# remove CarName
df_carPrices.drop(['CarName'], axis =1, inplace = True)
```

```python
df_carPrices.info()
```

```python
# Generating dummy values for categorical columns of more than 2 levels
df = pd.get_dummies(df_carPrices)
df.info()
```

```python
#defining a normalisation function 
cols_to_norm = ['symboling', 'carlength', 'carheight', 
         'enginesize', 'boreratio', 'stroke', 'compressionratio','horsepower', 'peakrpm', 'citympg', 'price']
# Normalising only the numeric fields 
normalised_df = df[cols_to_norm].apply(lambda x: (x-np.mean(x))/ (max(x) - min(x)))
normalised_df.head()

df['symboling'] = normalised_df['symboling']
df['carlength'] = normalised_df['carlength']
df['carheight'] = normalised_df['carheight']
df['enginesize'] = normalised_df['enginesize']
df['boreratio'] = normalised_df['boreratio']
df['stroke'] = normalised_df['stroke']
df['price'] = normalised_df['price']
df['compressionratio'] = normalised_df['compressionratio']
df['horsepower'] = normalised_df['horsepower']
df['peakrpm']= normalised_df['peakrpm']
df['citympg'] = normalised_df['citympg']
df.head()
```

```python
def vif_cal(input_data, dependent_col):
    vif_df = pd.DataFrame( columns = ['Var', 'Vif'])
    x_vars=input_data.drop([dependent_col], axis=1)
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]] 
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.OLS(y,x).fit().rsquared  
        vif=round(1/(1-rsq),2)
        vif_df.loc[i] = [xvar_names[i], vif]
    return vif_df.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False)
```
